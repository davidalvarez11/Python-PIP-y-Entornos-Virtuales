```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mfpass= pd.read_csv('Data/MfComparissionPass.csv')
del mfpass['playerid']
```

```python
mfpasstop50 = mfpass.head(50)

# Generar la grfica de barras
ax = mfpasstop50.groupby('club').size().plot(kind='bar', color='skyblue', edgecolor='black')

# Personalizar la grfica
ax.set_title('Distribucin del TOP50 MF', fontsize=15, fontweight='bold')
ax.set_xlabel('Club', fontsize=12)
ax.set_ylabel('Cantidad', fontsize=12)
ax.tick_params(axis='x', rotation=45)
ax.grid(True, linestyle='--', alpha=0.7)

# Mostrar la grfica
plt.tight_layout()
plt.show()
```

```python
# Generar la grfica de barras
ax = mfpasstop50.groupby('club').size().plot(kind='bar', color='skyblue', edgecolor='black')

# Personalizar la grfica
ax.set_title('Distribucin de Clubs en mfpasstop50', fontsize=15, fontweight='bold')
ax.set_xlabel('Club', fontsize=12)
ax.set_ylabel('Cantidad', fontsize=12)
ax.tick_params(axis='x', rotation=45)
ax.grid(True, linestyle='--', alpha=0.7)

# Agregar etiquetas dentro de las barras
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height() / 2),
                ha='center', va='center', fontsize=10, color='black', fontweight='bold')

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
